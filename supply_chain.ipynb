# Analyzing Demand and Supply
**About Dataset**

Understanding demand and supply is critical for businesses to optimize operations, maximize profits, and make informed decisions. Analyzing the demand and supply for cab ride businesses like Ola and Uber is one of the challenging use cases of demand and supply analysis. 

Cab services have become an essential part of urban transportation, with people relying heavily on these services for their daily commutes. Understanding the demand and supply patterns of cab services can help optimize their operations and provide a better user experience to customers.

* Drivers Active Per Hour: Number of drivers active per hour.
* Riders Active Per Hour: Number of Riders looking for rides.
* Rides Completed: Number of rides completed.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
df=pd.read_csv("dataset/rides.csv")
df=df.dropna()
df.head()
```

```python
print("Average Demand Per Hour: %.2f " % df["Riders Active Per Hour"].mean())
print("")
print("Average Supply Per Hour: %.2f "% df["Drivers Active Per Hour"].mean())
```

# Visualizing Supply and Demand Per Hour

```python
plt.scatter(df["Riders Active Per Hour"],df["Drivers Active Per Hour"])
plt.title("Demand and Supply Analysis")
plt.xlabel("Demand Per Hour")
plt.ylabel("Supply Per Hour")
plt.show()
```

# Finding Out and Visualize How Much of the Demand Did We Fullfill

```python
df["Demand Complete Ratio"]=(df["Rides Completed"]/df["Riders Active Per Hour"]*100)
df.head()
```

```python
plt.scatter(df["Riders Active Per Hour"],df["Demand Complete Ratio"])
plt.xlabel("Riders Active Per Hour (Demand)")
plt.ylabel("Completed Demand Ratio(%)")
plt.show()
```
